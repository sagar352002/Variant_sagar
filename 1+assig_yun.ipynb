```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.ril.com/businesses/energy/oil-gas-exploration-production'
```

```python


# Replace 'Your-User-Agent-Here' with the actual User-Agent from your browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36',
}

# Send a GET request with the custom headers
webpage = requests.get(url, headers=headers)
```

```python
# Parse the HTML content of the page
soup = BeautifulSoup(webpage.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# col-md-6 aos-init aos-animate
article = soup.find_all('div', class_ = "col-md-6 aos-init aos-animate")
```

```python
print(article)
```

```python
for i in article:
  print(i.text.strip())
```
